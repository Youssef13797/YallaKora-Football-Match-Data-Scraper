```python

all_matches = []

for chamionship in chamionships:
    
    championship_name = chamionship.find('h2').text.strip()
    matches = chamionship.find_all('div', {'class':'teamsData'})
    
    for match in matches:
        
        teamA = match.find('div', {'class':'teamA'}).find('p').text.strip()
        teamB = match.find('div', {'class':'teamB'}).find('p').text.strip()
        
        result_block = match.find('div', {'class':'MResult'})
        
        score_spans = result_block.find_all('span', {'class':'score'})
        
        scoreA = score_spans[0].text.strip()
        scoreB = score_spans[1].text.strip()
        
        time = result_block.find('span', {'class':'time'}).text.strip()
        
        match_data = {
            "Championship": championship_name,
            "Team A": teamA,
            "Team B": teamB,
            "Score A": scoreA,
            "Score B": scoreB,
            "Time": time
        }
        
        all_matches.append(match_data)

df = pd.DataFrame(all_matches)

df.head()
```
